# Câu 4 Kiểm tra bộ dữ liệu có bị khuyết. Thống kê số lượng giá trị khuyết từng biến

```python
#Xác định và thống kê số lượng giá trị bị khuyết
missing_values = [".", "?", "_", "-", "", "None", "NULL", "Not", "|", "l"]
def is_missing(value):
    return value in missing_values

missing_count = dataframe.map(is_missing).sum()

print(missing_count)
```

# Câu 5 Xuất ra các dòng bị khuyết. Lưu các kết quả bị khuyết

```python
# Lọc các dòng có giá trị bị khuyết
missing_rows = dataframe[dataframe.isin(missing_values).any(axis=1)]
# Xuất ra các dòng có giá trị bị khuyết
print(missing_rows)

# Lưu trữ kết quả vào tập tin .csv, .json, .xlsx
missing_rows.to_csv('missing_values.csv', index=False)
missing_rows.to_json('missing_values.json', orient='records', lines=True)
missing_rows.to_excel('missing_values.xlsx', index=False)
```

# Câu 6 Xử lý giá trị bị khuyết theo gợi ý

```python
#Tạo bản sao của dataframe và thay thế các giá trị bị khuyết thành np.NaN
df_copy_q6 = dataframe.copy()

df_copy_q6.replace(missing_values, np.nan, inplace=True)
df_copy_q6.apply(lambda row: np.nan if row.isnull().all() else row, axis=1)
```

```python
# Chuyển đổi kiểu dữ liệu của cột "stroke" sang float
df_copy_q6["stroke"] = df_copy_q6["stroke"].astype(float, errors='ignore')
# Thay thế các giá trị bị khuyết trong cột "stroke" bằng giá trị trung bình của cột
df_copy_q6["stroke"] = df_copy_q6["stroke"].fillna(df_copy_q6["stroke"].mean())
```

```python
#Thay thế các giá trị bị khuyết trong cột "normalized-losses" bằng giá trị xuất hiện nhiều nhất
df_copy_q6["normalized-losses"] = df_copy_q6["normalized-losses"].fillna(df_copy_q6["normalized-losses"].mode()[0])
```

```python
#Thay thế các giá trị bị khuyết trong cột "bore" bằng giá trị ở dòng trên
df_copy_q6["bore"] = df_copy_q6["bore"].ffill()
```

```python
#Thay thế các giá trị bị khuyết trong cột "horsepower" bằng giá trị ở dòng dưới
df_copy_q6["horsepower"] = df_copy_q6["horsepower"].bfill()
```

```python
#Chuyển kiểu dữ liệu của cột "peak-rpm" sang float
df_copy_q6["peak-rpm"] = df_copy_q6["peak-rpm"].astype(float, errors='ignore')
#Thay thế các giá trị bị khuyết trong cột "peak-rpm" bằng giá trị trung bình của cột
df_copy_q6["peak-rpm"] = df_copy_q6["peak-rpm"].fillna(df_copy_q6["peak-rpm"].mean())
```

```python
#Thay thế các giá trị bị khuyết trong cột "num-of-doors" bằng giá trị xuất hiện nhiều nhất
df_copy_q6["num-of-doors"] = df_copy_q6["num-of-doors"].fillna(df_copy_q6["num-of-doors"].mode()[0])
```

# Câu 7 Xử lý giá trị bị khuyết theo quy trình

```python
#Tạo bản sao của dataframe và thay thế các giá trị bị khuyết bằng np.NaN
df_copy_q7 = dataframe.copy()

df_copy_q7.replace(missing_values, np.nan, inplace=True)
df_copy_q7.apply(lambda row: np.nan if row.isnull().all() else row, axis=1)
```

```python
#Loại bỏ các dòng có giá trị bị khuyết của cột "price"
df_copy_q7.dropna(subset=["price"], inplace=True)
#Thay thế các giá trị bị khuyết trong cột "horse-power" bằng giá trị trung bình của cột
df_copy_q7["horsepower"] = df_copy_q7["horsepower"].astype(float, errors='ignore')
df_copy_q7["horsepower"] = df_copy_q7["horsepower"].fillna(df_copy_q7["horsepower"].mean())
#Thay thế các giá trị bị khuyết trong cột "normalized-losses" bằng giá trị xuất hiện nhiều nhất
df_copy_q7["normalized-losses"] = df_copy_q7["normalized-losses"].fillna(df_copy_q7["normalized-losses"].mode()[0])
#Loại bỏ tất cả các hàng có giá trị bị khuyết còn lại
df_copy_q7.dropna(inplace=True)
#Lưu trữ kết quả vào tập tin .xlsx
df_copy_q7.to_excel('df_copy_q7.xlsx', index=False)
```

# Câu 8 Dùng Nearest Neighbors để điền các giá trị khuyết

```python
from sklearn.impute import KNNImputer
```

```python
#Tạo bản sao của dataframe và thay thế các giá trị bị khuyết thành np.NaN
df_copy_q8 = dataframe.copy()

df_copy_q8.replace(missing_values, np.nan, inplace=True)
df_copy_q8.apply(lambda row: np.nan if row.isnull().all() else row, axis=1)
```

```python
#Xử lý giá trị kiểu số bị khuyết bằng nearest neighbors
# Chuyển đổi các cột cần thiết sang kiểu số
numeric_columns = ["normalized-losses", "bore", "stroke", "horsepower", "peak-rpm", "price"]
df_copy_q8[numeric_columns] = df_copy_q8[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Áp dụng Nearest Neighbors để xử lý giá trị bị khuyết
imputer = KNNImputer(n_neighbors=2)
df_copy_q8[numeric_columns] = imputer.fit_transform(df_copy_q8[numeric_columns])
```

# Câu 9. Tạo các bin cho các biến

```python
bin_labels = ['Low', 'Medium', 'High']

# Tạo bins cho "horsepower"
horsepower_bins = pd.cut(df_copy_q8['horsepower'], bins=3, labels=bin_labels)
df_copy_q8['horsepower_binned'] = horsepower_bins

# Tạo bins cho "price"
price_bins = pd.cut(df_copy_q8['price'], bins=3, labels=bin_labels)
df_copy_q8['price_binned'] = price_bins

# Tạo bins cho "city-mpg"
city_mpg_bins = pd.cut(df_copy_q8['city-mpg'], bins=3, labels=bin_labels)
df_copy_q8['city_mpg_binned'] = city_mpg_bins

# Tạo bins cho "highway-mpg"
highway_mpg_bins = pd.cut(df_copy_q8['highway-mpg'], bins=3, labels=bin_labels)
df_copy_q8['highway_mpg_binned'] = highway_mpg_bins

# In ra các cột và các bins tương ứng
df_copy_q8[['horsepower', 'horsepower_binned', 'price', 'price_binned', 'city-mpg', 'city_mpg_binned', 'highway-mpg', 'highway_mpg_binned']]
```

# Câu 10 Chuyển biến phân loại thành biến số bằng one-hot

```python
# Sử dụng phương pháp one-hot encoding
categorical_columns = ['make', 'fuel-type', 'aspiration', 'num-of-doors', 'body-style', 'drive-wheels', 'engine-location', 'engine-type', 'fuel-system']
df_one_hot_encoded = pd.get_dummies(df_copy_q8, columns=categorical_columns)

# In ra dữ liệu các cột đã được chuyển đổi:
df_one_hot_encoded.filter(like='make')
```

# Câu 11 Chuyển biến phân loại thành biến số bằng phương pháp khác

```python
# Chuyển các biến phân loại thành biến số bằng phương pháp frequency encoding
df_frequency_encoded = df_copy_q8.copy()
for col in categorical_columns:
    frequency = df_frequency_encoded[col].value_counts(normalize=True)
    df_frequency_encoded[col] = df_frequency_encoded[col].map(frequency)
    
# In ra dữ liệu các cột đã được chuyển đổi:
df_frequency_encoded[categorical_columns]
```

 # Câu 12 Chuẩn hóa dữ liệu bằng Min-Max

```python
# Chuẩn hóa dữ liệu bằng công thức Min-Max:
def min_max_scaling(series):
    return (series - series.min()) / (series.max() - series.min())

df_normalized_1 = df_copy_q8.copy()
numerical_columns = ['symboling', 'normalized-losses', 'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-size', 'bore', 'stroke', 'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']
for col in numerical_columns:
    df_normalized_1[col] = min_max_scaling(df_normalized_1[col])

# In ra dữ liệu các cột đã được chuẩn hóa:
df_normalized_1[numerical_columns]
```

```python
# Chuẩn hóa dữ liệu bằng module MinMaxScaler của sklearn:
from sklearn.preprocessing import MinMaxScaler

df_normalized_2 = df_copy_q8.copy()
scaler = MinMaxScaler()
df_normalized_2[numerical_columns] = scaler.fit_transform(df_normalized_2[numerical_columns])

# In ra dữ liệu các cột đã được chuẩn hóa:
df_normalized_2[numerical_columns]
```

# Câu 13 Chuẩn hóa dữ liệu bằng z-score

```python
# Chuẩn hóa dữ liệu bằng công thức Z-score:
def z_score_scaling(series):
    return (series - series.mean()) / np.std(series)

df_normalized_3 = df_copy_q8.copy()
for col in numerical_columns:
    df_normalized_3[col] = z_score_scaling(df_normalized_3[col])
    
# In ra dữ liệu các cột đã được chuẩn hóa:
df_normalized_3[numerical_columns]
```

```python
# Chuẩn hóa dữ liệu bằng module statmodels cho Z-score:
import scipy.stats
scipy.stats.zscore(df_copy_q8[numerical_columns])
```
