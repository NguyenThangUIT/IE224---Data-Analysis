```python
df = df.drop(columns=['User ID', 'Charging Station ID'])
df.head()
```

<p> Xử lý dữ liệu bị khuyết </p>

```python
# Kiểm tra tổng số null trong tất cả các feature
total_null = df.isnull().sum()
percent = df.isnull().sum() / df.shape[0] * 100.00
pd.concat([total_null, percent, df.dtypes], axis=1, keys=['Total', 'Percentage (%)', 'Type']).sort_values(by='Percentage (%)', ascending=False)
```

<p> Nhận xét: Dữ liệu ban đầu không bị khuyết quá nhiều </p>

```python
# Điền các giá trị khuyết bằng giá trị trung bình của từng cột
df['Distance Driven (since last charge) (km)'].fillna(df['Distance Driven (since last charge) (km)'].mean(), inplace=True)
df['Charging Rate (kW)'].fillna(df['Charging Rate (kW)'].mean(), inplace=True)

# Riêng cột mục tiêu thực hiện điền giá trị theo nhóm Vehicle Model, Battery Capacity (kWh), và Charger Type.
df['Energy Consumed (kWh)'] = df.groupby(
    ['Vehicle Model', 'Battery Capacity (kWh)', 'Charger Type']
)['Energy Consumed (kWh)'].transform(
    lambda x: x.fillna(x.mean())
)

# Với những nhóm không có giá trị để tính trung bình, điền bằng trung bình toàn tập
overall_mean = df['Energy Consumed (kWh)'].mean()
df['Energy Consumed (kWh)'] = df['Energy Consumed (kWh)'].fillna(overall_mean)

# Hiển thị số giá trị bị khuyết sau khi được điền để kiểm tra
df.isnull().sum()
```

<p> Chuyển đổi các cột có kiểu dữ liệu thời gian </p>

```python
# Chuyển đổi các cột thời gian sang định dạng datetime
df['Charging Start Time'] = pd.to_datetime(df['Charging Start Time'])
df['Charging End Time'] = pd.to_datetime(df['Charging End Time'])

# Trích xuất các thông tin thời gian
df['Charging Start Hour'] = df['Charging Start Time'].dt.hour
df['Charging End Hour'] = df['Charging End Time'].dt.hour
df['Charging Start Day'] = df['Charging Start Time'].dt.day
df['Charging End Day'] = df['Charging End Time'].dt.day
df['Charging Duration (hours)'] = (df['Charging End Time'] - df['Charging Start Time']).dt.total_seconds() / 3600

# Lọc các hàng có lỗi
invalid_time_rows = df[df['Charging End Time'] < df['Charging Start Time']]

# In số hàng lỗi để kiểm tra
print(f"Số hàng có lỗi thời gian: {len(invalid_time_rows)}")

# Kiểm tra kiểu dữ liệu sau khi chuyển đổi
print(df[['Charging Start Time', 'Charging End Time', 'Charging Duration (hours)']].dtypes)
```

```python
# Xác định các giá trị nhiễu dựa trên IQR (Interquartile Range)
def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Trả về chỉ số của các hàng chứa giá trị nhiễu
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)].index

# Xóa các giá trị nhiễu khỏi dataframe
def remove_outliers(df, numeric_columns):
    for col in numeric_columns:
        outlier_indices = detect_outliers(df, col)  # Lấy chỉ số hàng chứa giá trị nhiễu
        df = df.drop(index=outlier_indices)        # Loại bỏ các hàng nhiễu
    return df

# Chọn các cột số để phân tích
numeric_columns = df.select_dtypes(include=[np.number]).columns

# Loại bỏ giá trị nhiễu
df_cleaned = remove_outliers(df, numeric_columns)

# Kiểm tra lại số lượng giá trị nhiễu sau khi làm sạch
outliers_dict_after = {}
for col in numeric_columns:
    outliers_after = detect_outliers(df_cleaned, col)
    outliers_dict_after[col] = len(outliers_after)

# Hiển thị số lượng giá trị nhiễu còn lại
print("Số lượng giá trị nhiễu sau khi xóa:", outliers_dict_after)

# Vẫn còn 3 giá trị State of Charge (End %) chưa bị loại, nhóm vẫn chưa xác định được nguyên nhân
```

## Phân tích thăm dò và trực quan hóa dữ liệu

### Thống kê mô tả

<p> Central Tendency & Dispersion </p>

```python
# Chọn các cột số
numeric_columns = df.select_dtypes(include=[np.number])

# Mean
mean_values = numeric_columns.mean()

# Mode
mode_values = numeric_columns.mode().iloc[0]

# Median
median_values = numeric_columns.median()

# Range
range_values = numeric_columns.max() - numeric_columns.min()

# Quartiles
quartiles = numeric_columns.quantile([0.25, 0.5, 0.75])

# IQR
iqr_values = numeric_columns.quantile(0.75) - numeric_columns.quantile(0.25)

# Variance
variance_values = numeric_columns.var()

# Standard Deviation
std_values = numeric_columns.std()

# Coefficient of Variation (CV)
cv_values = std_values / mean_values

# Display the results
print("Mean:\n", mean_values)
print("\nMode:\n", mode_values)
print("\nMedian:\n", median_values)
print("\nRange:\n", range_values)
print("\nQuartiles:\n", quartiles)
print("\nIQR:\n", iqr_values)
print("\nVariance:\n", variance_values)
print("\nStandard Deviation:\n", std_values)
print("\nCoefficient of Variation (CV):\n", cv_values)
```

<p> Hình dạng của dữ liệu </p>

```python
from scipy.stats import skew, kurtosis
import matplotlib.pyplot as plt
import seaborn as sns

# Chọn các cột số
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Tính skewness và kurtosis cho từng cột
skewness = df[numerical_columns].apply(lambda x: skew(x.dropna()))
kurtosis_values = df[numerical_columns].apply(lambda x: kurtosis(x.dropna(), fisher=True))  # Fisher=True để trả về kurtosis chuẩn (trừ đi 3)

# Kết quả
result = pd.DataFrame({'Skewness': skewness, 'Kurtosis': kurtosis_values})
print(result)
```

<p> Hầu hết các cột có Skewness < |0.5|, cho thấy các phân phối khá đối xứng, không có độ lệch lớn. <br>
Battery Capacity (kWh): Skewness = 0.373 → Phân phối lệch phải nhẹ. <br>
State of Charge (End %): Skewness = 0.281 → Lệch phải nhẹ. <br>
Các cột khác có độ lệch nhỏ, cho thấy chúng rất đối xứng. </p>
<p> Hầu hết các cột có Kurtosis < 0, cho thấy dữ liệu phân tán rộng hơn phân phối chuẩn và ít tập trung quanh giá trị trung tâm. <br>
Các cột phẳng rõ rệt nhất:
Charging Duration (hours),
Vehicle Age (years),
Distance Driven (since last charge). <br>
Chỉ có hai cột có Kurtosis > 0:
Battery Capacity (kWh) (nhọn vừa phải).
State of Charge (End %) (nhọn hơn đáng kể).<br> </p>

### Phân tích ANOVA

```python
import seaborn as sns

import scipy.stats as stats
import matplotlib.pyplot as plt

# Lấy dữ liệu từ các biến phân loại
categorical_columns = df.select_dtypes(include=['object']).columns.to_list()

# Tạo lưới 3 cột 2 hàng
fig, axes = plt.subplots(2, 3, figsize=(18, 12))

# Vẽ các boxplot phân phối của các biến phân loại
for i, category in enumerate(categorical_columns):
    row = i // 3
    col = i % 3
    sns.boxplot(x=category, y='Energy Consumed (kWh)', data=df, ax=axes[row, col])
    axes[row, col].set_title(f'Boxplot of Energy Consumed (kWh) by {category}')
    axes[row, col].tick_params(axis='x', rotation=45)

# Điều chỉnh khoảng cách giữa các biểu đồ
plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

# Lưu kết quả
anova_results = []

# Lặp qua từng biến phân loại
for col in categorical_columns:
    # Tạo nhóm dữ liệu
    grouped_data = [df[df[col] == group]['Energy Consumed (kWh)'] for group in df[col].unique()]
    
    # Loại bỏ các nhóm trống hoặc giá trị khuyết
    grouped_data = [group.dropna() for group in grouped_data if not group.empty]
    
    # Thực hiện kiểm định ANOVA
    if len(grouped_data) > 1:  # Đảm bảo có ít nhất 2 nhóm để kiểm định
        f_statistic, p_value = stats.f_oneway(*grouped_data)
        anova_results.append((col, f_statistic, p_value))

# Chuyển kết quả thành DataFrame để dễ phân tích
anova_df = pd.DataFrame(anova_results, columns=['Variable', 'F-Statistic', 'P-Value'])

# In kết quả
print(anova_df.sort_values(by='P-Value'))
```

<p> Trong tất cả các biến phân loại, chỉ có Charger Type có F-Statistic đủ cao và P-Value < 0.05, cho thấy có sự khác biệt có ý nghĩa thống kê giữa trung bình Energy Consumed (kWh) của các nhóm trong biến Charger Type. Điều này cho thấy loại trạm sạc ảnh hưởng đến năng lượng tiêu thụ. </p>

```python
from geopy.geocoders import Nominatim
from tqdm import tqdm
import plotly.express as px

# Bước 1: Gom nhóm và đếm số lượng trạm sạc
location_data = df['Charging Station Location'].value_counts().reset_index()
location_data.columns = ['Charging Station Location', 'Number of Stations']

# Bước 2: Lấy tọa độ địa lý
geolocator = Nominatim(user_agent="geoapi_exercise")
tqdm.pandas()

def get_coordinates(location):
    try:
        loc = geolocator.geocode(location + ", USA")  # Giới hạn tìm kiếm trong nước Mỹ
        return (loc.latitude, loc.longitude)
    except:
        return (None, None)

# Áp dụng hàm lấy tọa độ
location_data['Coordinates'] = location_data['Charging Station Location'].progress_apply(get_coordinates)
location_data[['Latitude', 'Longitude']] = location_data['Coordinates'].apply(pd.Series)

# Bỏ các dòng không lấy được tọa độ
location_data = location_data.dropna(subset=['Latitude', 'Longitude'])

# Bước 3: Tạo biểu đồ địa lý với Plotly
fig = px.scatter_geo(location_data, 
                     lat='Latitude', 
                     lon='Longitude', 
                     size='Number of Stations', 
                     hover_name='Charging Station Location',
                     title="Phân phối số trạm sạc trên bản đồ Mỹ",
                     scope='usa',
                     color_discrete_sequence=['red'])

# Hiển thị biểu đồ
fig.show()
```

```python
import pandas as pd
import plotly.express as px

# Bước 1: Chuẩn bị dữ liệu
# Nhóm dữ liệu để tổng hợp năng lượng tiêu thụ
sunburst_data = df.groupby(['Charging Station Location', 'User Type', 'Charger Type'])['Energy Consumed (kWh)'].sum().reset_index()

# Bước 2: Tạo biểu đồ Sunburst
fig = px.sunburst(
    sunburst_data,
    path=['Charging Station Location', 'User Type', 'Charger Type'],  # Cấp bậc
    values='Energy Consumed (kWh)',  # Giá trị
    title="Phân tích tiêu thụ năng lượng theo 'Vị trí trạm sạc', 'Loại người dùng', 'Loại trạm sạc'",
    color='Energy Consumed (kWh)',  # Tô màu theo năng lượng tiêu thụ
    color_continuous_scale='spectral'  # Thang màu
)

# Bước 3: Hiển thị biểu đồ
fig.update_layout(margin=dict(t=40, l=0, r=0, b=0))
fig.show()
```
